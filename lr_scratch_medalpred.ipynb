#Linear Regression from Scratch

Predicting the no. of medals a country will win based on the number of athletes it enters into the olympics.

```python
import pandas as pd
import numpy as np
```

```python
teams=pd.read_csv('/content/drive/MyDrive/lr_scratch/teams.csv')
```

```python
teams #mainly- athletes, prev_medals
```

```python
X=teams[["athletes","prev_medals"]].copy()
y=teams[['medals']].copy()
```

```python
X
```

```python
y
```

```python
X["intercept"]=1
```

```python
X=X[["intercept","athletes","prev_medals"]]
```

```python
X
```

```python
X_T=X.T
```

```python
X_T
```

```python
B=np.linalg.inv(X_T@X)@X_T@y
```

```python
B
```

```python
B.index=X.columns
```

```python
B
```

```python
predictions=X@B
```

```python
predictions
```

```python
SSR=((y-predictions)**2).sum()
```

```python
SST=((y-y.mean())**2).sum()
```

```python
SSR
```

```python
SST
```

```python
R2=1-(SSR/SST)
```

```python
R2
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(teams[["athletes","prev_medals"]],teams[["medals"]])
```

```python
lr.intercept_
```

```python
lr.coef_
```
